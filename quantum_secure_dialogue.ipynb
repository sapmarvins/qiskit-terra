```python
import qiskit as qk
qc = qk.QuantumCircuit(12,12)
qc.x(range(12))
qc.h([0,2,4,6,8,10])
qc.cx([0,2,4,6,8,10], [1,3,5,7,9,11])
qc.barrier()
qc.measure(range(6,10),range(6,10))
qc.h([10,11])
qc.measure(range(10,12),range(10,12))
qc.reset(range(6,12))
qc.barrier()
qc.x(range(1))
qc.z(2)
qc.barrier()
qc.x(range(6,12))
qc.h([6,8,10])
qc.cx([6,8,10],[7,9,11])
qc.barrier()
qc.measure(range(4,6),range(4,6))
qc.reset(range(4,6))
qc.barrier()
qc.cx([0,2],[1,3])
qc.h([0,2])
qc.barrier()
qc.measure(range(0,4),range(0,4))
qc.reset(range(0,4))
qc.barrier()
qc.z(6)
qc.id(8)
qc.x(6)
qc.barrier()
qc.x(range(6))
qc.h([0,2,4])
qc.cx([0,2,4],[1,3,5])
qc.barrier()
qc.measure(range(10,12),range(10,12))
qc.reset(range(10,12))
qc.barrier()
qc.cx([6,8],[7,9])
qc.h([6,8])
qc.barrier()
qc.measure(range(6,10),range(6,10))
qc.reset(range(6,10))
qc.draw(output='mpl')
```
